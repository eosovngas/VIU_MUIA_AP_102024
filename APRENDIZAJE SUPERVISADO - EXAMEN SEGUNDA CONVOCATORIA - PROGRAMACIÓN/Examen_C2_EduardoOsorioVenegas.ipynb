#### 6) Comparación de modelos de regresión mediante validación cruzada (2.5 puntos)
Aplicar una validación cruzada interna de K=5 bolsas para optimizar y comparar la capacidad predictiva de los siguientes modelos: K-vecinos más cercanos (KNN) y Support Vector Machine (SVM). La comparación debe realizarse únicamente en términos de coeficiente de determinación (R2) proporcionando resultados de media +- desviación estándar.

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import make_scorer

# Definir los modelos
knn = KNeighborsRegressor()
svm = SVR()

# Crear un objeto KFold para la validación cruzada de 5 bolsas (K=5)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Validación cruzada para KNN utilizando R² como métrica
knn_r2_scores = cross_val_score(knn, X, y, cv=kf, scoring='r2')

# Validación cruzada para SVM utilizando R² como métrica
svm_r2_scores = cross_val_score(svm, X, y, cv=kf, scoring='r2')

# Calcular la media y desviación estándar del R² para cada modelo
knn_mean_r2 = np.mean(knn_r2_scores)
knn_std_r2 = np.std(knn_r2_scores)

svm_mean_r2 = np.mean(svm_r2_scores)
svm_std_r2 = np.std(svm_r2_scores)

# Mostrar los resultados
print(f"KNN - R2: {knn_mean_r2:.4f} ± {knn_std_r2:.4f}")
print(f"SVM - R2: {svm_mean_r2:.4f} ± {svm_std_r2:.4f}")
```

#### 7) Evaluación de los modelos sobre el conjunto de test (1 punto)
- Definir una función "plot_bisectriz()" cuyos inputs sean:
 * "y" --> la variable target,
 * "y_pred" --> las predicciones del modelo, y 
 * "name" --> el nombre del algoritmo utilizado.
 
- El output de dicha función será un gráfico basado en la bisectriz para ver las sobreestimaciones y subestimaciones de cada modelo

```python
import matplotlib.pyplot as plt

def plot_bisectriz(y, y_pred, name):
    plt.figure(figsize=(8, 6))
    
    # Graficar la bisectriz (línea donde y = y_pred)
    plt.plot([min(y), max(y)], [min(y), max(y)], color='gray', linestyle='--', label='Bisectriz')
    
    # Graficar los puntos (predicciones vs valores reales)
    plt.scatter(y, y_pred, color='blue', alpha=0.5, label='Predicciones')
    
    plt.xlabel('Valores reales (y)')
    plt.ylabel('Predicciones (y_pred)')
    plt.title(f'Comparación de predicciones vs reales ({name})')
    
    plt.legend()
    
    plt.grid(True)
    plt.show()
```

#### 8) Evaluación de los modelos sobre el conjunto de test (2 puntos)
- Entrenar los modelos anteriores utilizando todos los datos de entrenamiento.
- Evaluar su rendimiento sobre el conjunto de test.
- Llamar a la función "plot_bisectriz" para ver las sobreestimaciones y subestimaciones de cada modelo.
- Crear una tabla donde se muestren los resultados de todos los modelos.
 * Las filas serán: R2, MAE, MAPE, MSE y RMSE
 * Las columnas serán: KNN y SVM

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score

# Definir los modelos
knn = KNeighborsRegressor()
svm = SVR()

# Evaluando modelos con los X_train, X_test con la selección de caracteristicas, y_train, y_test con sus valores normalizados
X_train = X_train_selected
X_test = X_test_selected
y_train = y_train_scaled.ravel()
y_test = y_test_scaled.ravel()

# Entrenar los modelos con todos los datos de entrenamiento
knn.fit(X_train, y_train)
svm.fit(X_train, y_train)

# Predecir con los modelos en el conjunto de test
knn_pred = knn.predict(X_test)
svm_pred = svm.predict(X_test)

# Calcular las métricas para KNN
knn_r2 = r2_score(y_test, knn_pred)
knn_mae = mean_absolute_error(y_test, knn_pred)
knn_mape = mean_absolute_percentage_error(y_test, knn_pred)
knn_mse = mean_squared_error(y_test, knn_pred)
knn_rmse = np.sqrt(knn_mse)

# Calcular las métricas para SVM
svm_r2 = r2_score(y_test, svm_pred)
svm_mae = mean_absolute_error(y_test, svm_pred)
svm_mape = mean_absolute_percentage_error(y_test, svm_pred)
svm_mse = mean_squared_error(y_test, svm_pred)
svm_rmse = np.sqrt(svm_mse)
```

```python
# Llamar a la función "plot_bisectriz" para visualizar las sobreestimaciones y subestimaciones
plot_bisectriz(y_test, knn_pred, name="KNN")
plot_bisectriz(y_test, svm_pred, name="SVM")
```

```python
# Crear una tabla con los resultados
results = pd.DataFrame({
    'KNN': [knn_r2, knn_mae, knn_mape, knn_mse, knn_rmse],
    'SVM': [svm_r2, svm_mae, svm_mape, svm_mse, svm_rmse]
}, index=['R2', 'MAE', 'MAPE', 'MSE', 'RMSE'])

# Mostrar la tabla de resultados
results
```
